```python
import pandas as pd
import seaborn as sns
sns.set()

iris=sns.load_dataset("iris")

#csvで読み込む場合
#file=pd.read_csv("XXXX.csv)
#sns.pairplot(iris)

#棒グラフ
sns.catplot(x="petal_length",data=iris,kind="count")
#棒グラフにエラーバーをつける
sns.catplot(x="petal_length",y="species",data=iris,kind="bar")
```
